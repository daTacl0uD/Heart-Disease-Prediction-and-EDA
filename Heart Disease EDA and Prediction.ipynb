```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
Y_pred=model.predict(predictor_test)
accuracy = accuracy_score(target_test,Y_pred)
accuracy
```

## Checking Confusion Matrix & Accuracy, Precision and Recall Score

```python
from sklearn.metrics import confusion_matrix
cmatrix = confusion_matrix(target_test.tolist(),Y_pred.tolist())
cmatrix
print("Accuracy:",accuracy_score(target_test,Y_pred))
print("Precision Score :",precision_score(target_test,Y_pred))
print("Recall Score :",recall_score(target_test,Y_pred))
print("f1 Score :",f1_score(target_test,Y_pred))
print("Confusion Matrix :\n", cmatrix)
print('Sensitivity : ', cmatrix[0,0]/(cmatrix[0,0]+cmatrix[1,0]))
print('Specificity : ', cmatrix[1,1]/(cmatrix[1,1]+cmatrix[0,1]))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
#Get predicted probabilites from the model
y_probabilities0 = model.predict_proba(predictor_test)[:,1]
```

```python
#Create true and false positive rates
false_positive_rate_RF,true_positive_rate_RF,threshold_RF = roc_curve(target_test,y_probabilities0)
```

## Plotting ROC Curve

```python
#Plot ROC Curve
plot.pyplot.figure(figsize=(10,6))
plot.pyplot.title('ROC Curve')
plot.pyplot.plot(false_positive_rate_RF,true_positive_rate_RF)
plot.pyplot.plot([0,1],ls='--')
plot.pyplot.plot([0,0],[1,0],c='.5')
plot.pyplot.plot([1,1],c='.5')
plot.pyplot.xlabel('False Positive Rate')
plot.pyplot.ylabel('True Positive Rate')
plot.pyplot.show()
```

```python
#Calculate area under the curve
roc_auc_score(target_test,y_probabilities0)
```
